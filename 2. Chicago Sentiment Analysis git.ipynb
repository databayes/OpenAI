```python
import openai
import pandas as pd

# Azure OpenAI configuration
openai.api_type = "azure"
openai.api_base = "https://sueopenai.openai.azure.com/"  # replace with yours
openai.api_version = "2023-07-01-preview"  # or your API version
openai.api_key = openapiKey

# Your deployed model name (deployment name, not model name)
DEPLOYMENT_NAME = "gpt-35-turbo"  # for example
```

## üì• Load the review dataset

```python

# Load data into pandas DataFrame from "/lakehouse/default/Files/demo_reviews.csv"
df = pd.read_csv("/lakehouse/default/Files/chicago_reviews_full.csv")
df['length'] = df['review_text'].apply(len)
display(df)
df.shape
```

## üß™ Demo 1: Summarise a single review

```python
def summarise_review(text):
    response = openai.ChatCompletion.create(
        engine= DEPLOYMENT_NAME,
        messages=[{"role": "user", "content": f"Summarise this review: {text}"}]
    )
    return response['choices'][0]['message']['content']

# Example:
print(summarise_review(df['review_text'][0]))
```

```python
# Example usage (load from your demo CSV or inline)
df['summary'] = df['review_text'].apply(summarise_review)
df[['review_text', 'summary']]
```

## üé≠ Demo 2: Classify sentiment

```python
def classify_sentiment(text):
    response = openai.ChatCompletion.create(
        engine=DEPLOYMENT_NAME,
        messages=[
            {"role": "user", "content": f"Classify this review as Positive, Neutral, or Negative: {text}"}
        ]
    )
    return response['choices'][0]['message']['content']
```

```python
# Example usage (load from your demo CSV or inline)
df['sentiment'] = df['review_text'].apply(classify_sentiment)
df[['review_text', 'sentiment']]
```

## üîç Demo 3: Extract topics or key issues

```python
def extract_topic(text):
    response = openai.ChatCompletion.create(
        engine=DEPLOYMENT_NAME,
        messages=[
            {"role": "user", "content": f"List one topic mentioned in this review: {text}"}
        ]
    )
    return response['choices'][0]['message']['content']
```

```python
# Example usage (load from your demo CSV or inline)
df['topic'] = df['review_text'].apply(extract_topic)
display(df)
```

```python
df[['review_text', 'sentiment', 'topic']]
```

```python
def extract_topics(text):
    try:
        response = openai.ChatCompletion.create(
            engine=DEPLOYMENT_NAME,
            messages=[
                {"role": "user", "content": f"List the main topics mentioned in this review: {text}"}
            ]
        )
        # Safely extract content if available
        choices = response.get('choices', [])
        if choices and 'message' in choices[0] and 'content' in choices[0]['message']:
            return choices[0]['message']['content']
        else:
            print("‚ö†Ô∏è Missing 'message.content' in response:", response)
            return "No content returned"
    except Exception as e:
        print(f"‚ùå Exception for input: {text}\n", e)
        return "ERROR"
```

## üîç Demo 4: Suggest Action

```python
def suggest_action(text):
    response = openai.ChatCompletion.create(
        engine=DEPLOYMENT_NAME,
        messages=[{"role": "user", "content": f"Suggest one concise actionable improvement based on this review: {text}"}]
    )
    return response['choices'][0]['message']['content']

df['recommendation'] = df['review_text'].apply(suggest_action)
df[['review_text', 'recommendation']]
display(df)
```
